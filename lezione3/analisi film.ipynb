```python
import pandas as pd
```

Leggiamo i file, in modo che ognuno sia in un dataframe separato.

```python
film = pd.read_csv("data/film.csv")
film.head()
```

```python
ruoli = pd.read_csv("data/ruoli.csv")
ruoli.head()
```

```python
produzione = pd.read_csv("data/produzione.csv")
produzione.head()
```

## 1

Per ogni film, calcolare il numero di componenti del cast e mostrare anche il titolo.

```python
quanti_cast = ruoli.groupby(by = 'id_film', as_index = False).count()[['id_film', 'personaggio']]
quanti_cast.head()
```

```python
pd.merge(film, quanti_cast, left_on = 'id', right_on = 'id_film')[['id', 'title', 'personaggio']]
```

## 2

Per ogni anno, determinare quanti film non hanno una homepage.

```python
no_homepage = film[film["homepage"].isnull()].groupby('anno').size()
no_homepage.head()
```

## 3

Per ogni film, calcolare il margine operativo lordo -- la differenza fra incassi (revenue) e spese (budget)

```python
film['margine'] = film['incassi'] - film['spese']
```

```python
film[['id', 'title', 'incassi', 'spese', 'margine']]
```

## 4

Per ogni società (company) e ogni decennio, calcolare gli incassi totali

```python
film['decennio'] = film['anno'] // 10
film.head()
```

```python
pd.merge(film, produzione, left_on = 'id', right_on = 'id_movie').groupby(['id_y', 'name', 'decennio']).sum()['incassi']
```

## 5

Per ogni decennio, determinare la società che ha realizzato gli incassi totali più elevati (in caso di parità, è sufficiente mostrarne una).

```python
gruppi = pd.merge(film, produzione, left_on = 'id', right_on = 'id_movie').groupby(['id_y', 'name', 'decennio'], as_index = False).sum()[['id_y', 'name', 'decennio', 'incassi']]
gruppi.head()
```

```python
migliori = gruppi.groupby('decennio', as_index = False).idxmax()['incassi']
migliori
```

```python
gruppi.iloc[migliori]
```

## 6

Trovare i film in cui un attore o un'attrice ha interpretato due personaggi diversi

```python
num_ruoli = ruoli.groupby(['id_film', 'id_attore'], as_index = False).count()[['id_film', 'attore']]
num_ruoli.head()
```

```python
pd.merge(film, num_ruoli[num_ruoli['attore'] > 1], left_on = 'id', right_on = 'id_film')[['id_film', 'title', 'attore']]
```
