```python
import pandas as pd
```

```python
film = pd.read_csv('film.csv')
produzione = pd.read_csv('produzione.csv')
ruoli = pd.read_csv('ruoli.csv')
```

#### Per ogni film, calcolare il numero di componenti del cast e mostrare anche il titolo.

```python
#Le informazioni sono nel dataset film e ruoli: uniamoli specificando le colonne corrispondenti
film_ruoli = pd.merge(film, ruoli, left_on='id', right_on='id_film')

#Poi raggruppiamo sul titolo e usiamo l'operatore count() per contare quante righe sono associate ad ogni film (quindi quanti ruoli)
film_ruoli.groupby('title')['title'].count()
```

#### Per ogni anno, determinare quanti film non hanno una homepage

```python
#In questo caso ci serve solo il database film. 
#Prima creiamo in dataset con quelli che non hanno homepage 
na_homepage = film[film['homepage'].isnull()]

#Poi raggruppiamo sull'anno
na_homepage.groupby('anno')['anno'].count()
```

#### Per ogni film, calcolare il margine operativo lordo --- la differenza fra incassi (revenue) e spese (budget)

```python
#Anche in questo caso ci serve solo il dataset film
film['margine'] = film['incassi'] - film['spese']

#visualizziamo il risultato
film[['title', 'margine']]
```

#### Per ogni società (company) e ogni decennio, calcolare gli incassi totali

```python
#In questo caso ci serviranno i dataset film (per gli incassi) e produzione (per le company)
film_prod = pd.merge(film, produzione, left_on='id', right_on='id_movie')

#Creiamo il range decennale
decade = 10 * (film_prod['anno'] // 10)
decade.name = 'decade'

#Raggruppiamo sul nome della compagnia e per decennio
film_prod.groupby(['name', decade])['incassi'].sum()
```

```python
#Alternativamente, per visualizzare l'output in formato esteso
film_prod.groupby(['name', decade])['incassi'].sum().unstack().fillna(0)
```

#### Per ogni decennio, determinare la società che ha realizzato gli incassi totali più elevati (in caso di parità, è sufficiente mostrarne una)

```python
#Prima di tutto raggruppiamo e calcoliamo la somma degli incassi
sommaIncassi = pd.DataFrame(film_prod.groupby([decade, 'name'])['incassi'].sum())

#Poi troviamo la riga corrispondente al massimo degli incassi, sempre diviso per decade
idx = sommaIncassi.groupby(['decade'])['incassi'].idxmax()

#Infine usiamo loc per ottenere le righe corrispondenti
sommaIncassi.loc[idx]
```

#### Trovare i film in cui un attore o un'attrice ha interpretato due personaggi diversi

```python
#Per questo esercizio useremo il film e ruoli (abbiamo già il dataset unito nell'esercizio 1)
#Questa funzione usa filter per estrarre quei film che hanno un attore in due ruoli
film_trovati = film_ruoli.groupby(['title', 'id_attore']).filter(lambda x: x['ruolo_in_film_id'].count() == 2)['title']

#Il risultato dell'operazione contiene i titoli del film ripetuti, quindi generiamo un database e poi manteniamo solo un titolo
film_trovati = pd.DataFrame(film_trovati)
film_trovati.groupby('title').first()
```
